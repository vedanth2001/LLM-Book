```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=False,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

AutoModelForCausalLM.from_pretrained() loads the pretrained Phi-3-mini-4k-instruct causal language model (a GPT-style next-token prediction model) along with its learned weights and configuration from Hugging Face. The parameter device_map="cuda" places the model on the GPU for faster computation (if a GPU is available), while torch_dtype="auto" automatically selects an appropriate numerical precision (such as float16 on GPU) to optimize memory usage and speed. The parameter trust_remote_code=False prevents execution of custom model code from the repository and restricts loading to built-in Transformers implementations for safety. AutoTokenizer.from_pretrained() loads the matching tokenizer, which converts input text into token IDs that the model can process and later decodes generated token IDs back into human-readable text.`
